```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

% matplotlib inline
```

```python
src = cv2.imread('images/morphological.png')
plt.imshow(src)
```

```python
kernel = np.ones((5,5),np.uint8)
# 腐蚀操作是对高亮区域的腐蚀
erosion = cv2.erode(src, kernel, iterations=1)
plt.imshow(erosion)
```

```python
# 膨胀是对高亮区域的膨胀
dilation = cv2.dilate(src, kernel, iterations=1)
plt.imshow(dilation)
```

## 开运算，先腐蚀再膨胀，可以消除值比较高的噪音

```python
# 开运算：先腐蚀再膨胀，可以消除值比较高的噪音
noise1 = cv2.imread('images/noise1.png')

kernel = np.ones((7,7), np.uint8)
des1 = cv2.morphologyEx(noise1, cv2.MORPH_OPEN, kernel)
plt.subplot(121)
plt.imshow(noise1)
plt.subplot(122)
plt.imshow(des1)
```

## 闭运算，先膨胀再腐蚀，可以消除值比较低的噪音，即填满空洞

```python
noise2 = cv2.imread('images/noise2.png')

des2 = cv2.morphologyEx(noise2, cv2.MORPH_CLOSE, kernel)

plt.subplot(121)
plt.imshow(noise2)
plt.subplot(122)
plt.imshow(des2)
```
